```python
def lcm_of(x,y):
    
    if x>y:
        bigger = x
    else:
        bigger = y
        
    while(True):
        if((bigger%x==0) and (bigger%y==0)):
            lcm=bigger
            break
        bigger +=1
    
    return lcm
    
num1 = 54
num2 = 24

print("The L.C.M. is",lcm_of(num1, num2))
```

```python
#5.	Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?
```

```python
def add(x,y):
    return x+y

def subtract(x,y):
    return x-y

def multiply(x,y):
    return x*y

def divide(x,y):
    return x/y

print("Enter Choice")
print("1. Addition ")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

while(True):
    choice = input("Enter your choice")
    
    if choice in("1","2","3","4"):
        num1=int(input("Enter First Number "))
        num2=int(input("Enter Second Number "))
        
        if choice=="1":
            print(num1, "+" ,num2 , "=" , add(num1,num2))
            
        elif choice=="2":
            print(num1, "-",num2 , "=" , subtract(num1,num2))
            
        elif choice=="3":
            print(num1, "*", num2 , "=",  multiply(num1,num2))
            
        elif choice=="4":
            print(num1, "/", num2 , "=", divide(num1,num2))
            
            
            
    else:
        print("Invalid Input")
        break
```

```python
#1.	Write a Python Program to Add Two Matrices?
```

```python
import numpy as np

a=np.array([[1,2,3],[3,4,5]])
b=np.array([[3,4,5],[2,4,7]])
c=np.add(a,b)
print("Addition of two matrices:", c)
```

```python
#2.	Write a Python Program to Multiply Two Matrices?
```

```python
import numpy as np

a=np.array([[2,3,3],[3,2,4]])
b=np.array([[3,5],[2,3],[2,3]])
c=a.dot(b)
print("Multiplication of two arrays are :",c)
```

```python
#5.	Write a Python Program to Remove Punctuation From a String?
```

```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

a = "HI!!! HELLO[][][]() THANK YOU...."

b= ""
for char in a:
    if char not in punctuations:
        b = b + char

print(b)
```

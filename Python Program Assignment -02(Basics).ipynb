```python
km=int(input("enter kms = "))
miles = km*0.621371 #CONVERSION OF KMS TO MILES
print("miles =",miles)
```

```python
#2.	Write a Python program to convert Celsius to Fahrenheit?
```

```python
c=int(input("Enter Celcius = "))
Fahrenheit= (c*9/5)+32 #CONVERSION OF CELCIUS TO FAHRENHEIT
print("Fahrenheit is ", Fahrenheit)
```

```python
#4.	Write a Python program to solve quadratic equation?
```

```python

import cmath

a = 1
b = 2
c = 4


d = (b**2) - (4*a*c)          #DISCRIMINANT OF THE EQUATION

x = (-b-cmath.sqrt(d))/(2*a)  #SOLUTIONS OF THE EQUATION 
y = (-b+cmath.sqrt(d))/(2*a)

print('The solution is x & y ',x,y)
```

```python
#5.	Write a Python program to swap two variables without temp variable?
```

```python

x = 10
y = 9

print("Value of x : ", x, " and y : ", y)

x, y = y, x #SWAPPING WITHOUT TEMP VARIABLE 

print("Value of x : ", x, " and y : ", y)
```
